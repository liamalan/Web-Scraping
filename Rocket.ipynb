```python
def grab(company_name):
    print(company_name)
    driver.get('https://www.google.com')
    print(' Find search bar')
    time.sleep(2)
    search_bar = driver.find_element_by_name('q')
    print(' Enter search term')
    time.sleep(2)
    search_bar.send_keys('rocketreach ' + company_name)
    search_bar.send_keys(Keys.RETURN)

    time.sleep(2)
    print(" Find first link")
    try:
        first_link = driver.find_element_by_css_selector("#rso > div:nth-child(1) > div > div > div.yuRUbf > a > h3 > span")
    except NoSuchElementException:
        first_link = driver.find_element_by_css_selector("#rso > div:nth-child(1) > div > div.yuRUbf > a > h3 > span")
    first_link.click()
    
    time.sleep(5)
    print(" Find first combo")
    first_combo = driver.find_element_by_css_selector("#blur-container > div.rr-content > div > div.rr-section.rr-sec-card.container > div > div > div > div > div > div.table-wpr.gutter-padding > table > tbody > tr:nth-child(1) > td:nth-child(1)")

    print(" Find first email")
    first_email  = driver.find_element_by_css_selector("#blur-container > div.rr-content > div > div.rr-section.rr-sec-card.container > div > div > div > div > div > div.table-wpr.gutter-padding > table > tbody > tr:nth-child(1) > td:nth-child(2)")
    
    return([first_combo.text, first_email.text])
    
```
